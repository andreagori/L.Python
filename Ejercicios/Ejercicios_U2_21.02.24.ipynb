```python
# Ejercicio 10
"""
Escriba un script(guión) que ingrese tres números enteros desde el
usuario. Muestre la suma, el promedio, el producto, el menor y el
mayor de los 3 números. Tenga en cuenta que cada uno de estos es
una reducción en la programación de estilo funcional.

Pedir usuario que ingrese tres numeros, sumarlos, promedio, producto, mayor, menor.
"""

numUno = int(input('Ingresa el primer numero: \n'))
numDos = int(input('Ingresa el segundo numero: \n'))
numTres = int(input('Ingresa el tercer numero: \n'))
mayor = numUno
menor = numTres

if mayor < numDos:
    mayor = numDos
if mayor < numTres:
    mayor = numTres
    
if menor > numDos:
    menor = numDos
if menor > numUno:
    menor = numUno

print('La suma de los numeros (',numUno, numDos, numTres,') es:', numUno + numDos + numTres, '\n')
print('El promedio de los numeros (',numUno, numDos, numTres,') es:', (numUno + numDos + numTres) // 3, '\n')
print('El producto de los numeros (',numUno, numDos, numTres,') es:', numUno * numDos * numTres, '\n')
print('El mayor de los numeros (',numUno, numDos, numTres,') es:', mayor, '\n')
print('El menor de los numeros (',numUno, numDos, numTres,') es:', menor, '\n')
```
